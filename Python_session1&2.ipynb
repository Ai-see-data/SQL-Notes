# Data Types in Python
Python supports several built-in data types:
Integers (int): Whole numbers (e.g., 10 , -5).
Floats (float): Decimal numbers (e.g., 3.14 , -0.001).
Strings (str): Text data enclosed in quotes (e.g., "Hello" , 'Python').
Booleans (bool): Represents True or False .
Lists: Ordered, mutable collections (e.g., [1, 2, 3]).
Tuples: Ordered, immutable collections (e.g., (1, 2, 3)).
Sets: Unordered collections of unique elements (e.g., {1, 2, 3}).
Dictionaries: Key-value pairs (e.g., {"name": "Alice", "age": 25})

```python
#Use the type() function to check the data type of a variable
print(type(a)) 
print(type(height))
print(type("Aish"))
```

```python
#Typecasting 
#process of converting one data type to another.
#int() : Converts to integer.
#float() : Converts to float.
#str() : Converts to string.
#bool() : Converts to boolean.
# Convert string to integer and input function
num =str(input("Enter a number"))
print(type(num))
num =int(num)
print(num)
print(type(num))
```

```python
#Arithmetic Operators
print(10+5, 10-5, 10*2, 10/2, 10%2)
#Comparison Operators
print(10==5, 10==10, 10>5, 10<5)
#Logical Operators
print(True and False)
print(True and True)
#membership ( in and not in)
ls = ["a", "b", "c"]
print("b" in ls) 
print("d" in ls)
#identity ( is, is not)
x = 10
y = 10
z = 5
print(x is z)
```

```python
#Branching
#Conditional statements allow you to execute code based on certain conditions.
#Python uses if , elif , and else for decision-making
age = int(input("enter your age:"))
if age <18:
    print("You are not eligible.")
elif age ==18:
    print("waiting list")
else:
    print("You are eligible.")
```

```python
#for loop
#for loops are used to iterate over a sequence (e.g., list, string, range).
#They execute a block of code repeatedly for each item in the sequence.
for i in range(5):
    print(i)
```

```python
ls=[6,5,4,3,2,1]
for i in ls:
    a='*'
    b=a*i
    print(b)
```

```python
#break 
#exits the loop when a condition when is satisfied
for i in range(10):
    if i == 5: #koop breaks here and does not continue further
        break
    print(i)
```

```python
#continue
#The continue statement skips the rest of the code in the current iteration and moves to the next iteration.
for i in range(10):
    if i == 5: #loop skips this value and continues further
        continue
    print(i)
```

```python
#While loop
#While loops execute a block of code as long as a condition is True.
#They are useful when the number of iterations is not known in advance.
count = 0
while count < 5:
    print(count)
    count += 1
```

```python
#Strings - these are immutable
#A string is a sequence of characters enclosed within either single quotes,double quotes or triple quotes
#string indexing
text = "Hello, I'm Aish"
print(len(text))
print(text[0])
print(text[1]) 
print(text[-1]) 

#slicing
print(text[0:5]) 
print(text[:5]) 
print(text[7:]) 
print(text[::2]) #step size is included
print(text[::-1]) #reversing string
```

```python
#string methods
print(text.upper()) 
print(text.lower()) 
print(text.find('l'))

txt='Assignment'
print(txt.split('m'))
print(txt.partition('m'))
```

```python
#lists
#Lists are ordered, mutable (changeable) collections of items
#used for storing and manipulating data
#nos =[1,2,3,4,5]
#val =[10,"hello",3.14]
my_list =[1,2,3]
my_list.append(4) 
print(my_list)
my_list.insert(1,99) 
print(my_list)
my_list.remove(2) 
print(my_list)
my_list.pop()
print(my_list)
my_list.reverse() 
print(my_list)
my_list.sort()
print(my_list)
```
